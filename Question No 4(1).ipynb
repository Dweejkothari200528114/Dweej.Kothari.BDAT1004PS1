# Question No 4
Write a program in C# Sharp to separate odd and even integers in separate arrays. 
Go to the editor
Test Data :
Input the number of elements to be stored in the array :5 
Input 5 elements in the array :
element - 0 : 25
element - 1 : 47
element - 2 : 42
element - 3 : 56
element - 4 : 32 
Expected Output:
The Even elements are: 
42 56 32
The Odd elements are : 
25 47

```
using System;  
public class OddEven
{  
    public static void Main() 
 {
    int[] arr1 = new int[10];
    int[] arr2 = new int[10];
    int[] arr3 = new int[10];
    int d,x=0,y=0,n;
    
    
       Console.Write("\n\nSeparate odd and even integers in diff arrays:\n");
       Console.Write("------------------------------------------------------\n");

       Console.Write("Input number of elements to be stored in array :");
       n = Convert.ToInt32(Console.ReadLine());    
   
       Console.Write("Input {0} elements in the array :\n",n);
       for(d=0;d<n;d++)
            {
               Console.Write("element - {0} : ",d);
              arr1[d] = Convert.ToInt32(Console.ReadLine());  
            }

        for(d=0;d<n;d++)
        {
            if (arr1[d]%2 == 0)
                {
                   arr2[x] = arr1[d];
                   x++;
                }
            else
            {
                arr3[y] = arr1[d];
                y++;
            }
        }

    Console.Write("\nThe Even elements are : \n");
    for(d=0;d<x;d++)
    {
        Console.Write("{0} ",arr2[d]);
    }

        Console.Write("\nThe Odd elements are :\n");
        for(d=0;d<y;d++)
        {
            Console.Write("{0} ", arr3[d]);
        }
            Console.Write("\n\n");
 }
}
```
